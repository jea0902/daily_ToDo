```python
# 뉴스 크롤링

import urllib.request as req
from bs4 import BeautifulSoup

url = "http://media.daum.net"

# url 요청
res = req.urlopen(url)
source = res.read()

# 소스 디코딩
source = source.decode('utf-8')

# html 파싱
html = BeautifulSoup(source, 'html.parser')

# tab[속성=값] 요소 추출
atags = html.select('a[class=link_txt]') # 리스트

# a 태그 내용 수집
crawling_data=[]

cnt = 0
for atag in atags:
    cnt += 1
    atagStr = str(atag.string)
    
    crawling_data.append(atagStr.strip())
    
# 수집한 자료 확인
print('crawling result')
print(type(crawling_data)) # '리스트'

# text save : object -> file(string) -> load(string)
file = open('./crawling_data.txt',mode='w',encoding='utf-8')
file.write(str(crawling_data)) # save
file.close()
print('text file 확인')
```
