```python
from sklearn.datasets import load_iris

import pandas as pd
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
iris = load_iris()
```

```python
iris.data # 크게보면 리스트 1 - 작은 리스트들 (4개열 쓸 수 있는)
```

```python
iris # 자세히 보니, data 딕셔, target 딕셔 두개로 이뤄짐.
```

```python
iris.target
```

```python
col_names = [
             'sepal_length',  # 꽃받침의 길이
             'sepal_width' ,  # 꽃받침의 폭
             'petal_length',  # 꽃잎의 길이
             'petal_width' ,  # 꽃잎의 폭
             'target'
            ]
```

```python
iris_df = pd.DataFrame(data = np.c_[iris.data,iris.target], columns=col_names)
iris_df.target = iris_df.target.map({0:'setosa',1:'versicolor', 2:'virginica'})
```

```python
iris_df
```

```python
plt.figure(figsize=(8,8))
for i in range(0,4):
    plt.subplot(2,2,i+1)
    sns.distplot(iris_df.iloc[:,i]) # 열마다 히스토
plt.tight_layout()
plt.show()
```

```python
# 상관계수 함수 : 두 변수의 선형 상관계수.
# col1이 증가할 경우 col2는 대체로 증가 => 거의 선형

iris_corr = iris_df.corr()
iris_corr
```

```python
sns.pairplot(iris_df) # pairplot 여러 변수간 산점도
plt.show()
```

```python
# 산점도에 줄을 그어라

sns.pairplot(iris_df, kind='reg')
plt.show()
```

```python
plt.figure(figsize=(8,8))

# 히트맵 : 열을 의미하는 heat와 지도를 뜻하는 map
# 데이터들의 배열을 색상으로 표현해주는 그래프.

# 두개의 카테고리 값에 대한 값 변화를 한 눈에 알기 쉽다.

# annot = True => 각 셀에 숫자를 입력
# fmt = '.2f' -> float인데, 0.2배로 써라.
# linewidths는 각 셀 사이마다 선을 넣어주고, 그 선의 굵기를
# 정할 수 있다.

# mask = np.array(iris_corr)
# mask[np.tril_indices_from(mask)] = False

# mask 파라미터 추가
sns.heatmap(iris_corr,annot=True,fmt='.2f'
            ,linewidths=.5
            ,cmap='Blues')
plt.show()
```
