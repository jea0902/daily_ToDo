BeautifulSoup : 
html 코드를 Python이 이해하는 객체 구조로 변환하는 Parsing을 하는 라이브러리

```python
import urllib.request # 원격 서버 파일 요청
from bs4 import BeautifulSoup # html 파싱
# 파싱은 데이터를 분해,분석하여 원하는 형태로 조립하고,
# 다시 빼내는 프로그램.

# 요청할 url 
url = 'http://www.naver.com/index.html'
```

```python
# 1 원격 서버 파일 요청
res = urllib.request.urlopen(url) # web 문서 get
                            # 웹 리소스를 가져오는 방법.

data = res.read() # text 형태로 읽음 - 가져온 것.
print(data)
```

```python
# 2 source 디코딩 : 한글로 디코딩
src = data.decode("utf-8")
print('source')
print(src)
```

```python
# html 파싱
# html 태그를 인식할 수 있도록 의미있는 단위로 분해
# 계층적인 트리 구조를 만드는 과정

html = BeautifulSoup(src,'html.parser')
print('html 파싱')
print(html)
```

```python
# 4. 태그 내용
a = html.find('a')
print(a)
print(a.string)
```
